# 수행평가 결과 입력하기
##1. 환경 설정, 학생 입력(csv)
##2. 과목별 기준, 평가단계(3~5단계) 입력하기
##3. 과목별 기본점수 입력
### 3-1. 학생 점수를 입력하면 평가결과 알려주기
##4. 표로 나타내기

```python
#폰트 설정
!apt-get install fonts-nanum -qq > /dev/null
!fc-cache -fv
import matplotlib as mpl
mpl.font_manager._rebuild()
findfont = mpl.font_manager.fontManager.findfont
mpl.font_manager.findfont = findfont
mpl.backends.backend_agg.findfont = findfont

#환경 설정
import csv
import matplotlib.pyplot as plt
import time
import pandas as pd
student=[]
subject=[]
score=[]
result=[]


#학생 명단 불러오기(+리스트 속 리스트 만들기)
data = open('5~6주차 과제용(학생 명단).txt')
for row in data:
  student.append(row.strip())
print(student)

for i in range(len(student)):
  score.append([])
  result.append([])
print(score)
print(result)

#과목 입력
answer = input('평가를 실시하는 과목의 단원명을 1가지 적으세요.')
subject.append(answer)
passing = input('과목을 더 입력하시겠습니까?( y / n )')
while True: 
  if passing == 'y' :
    answer_plus = input('평가를 실시하는 과목의 단원명을 1가지 적으세요.')
    subject.append(answer_plus)
    passing = input('과목을 더 입력하시겠습니까?( y / n )') 
  elif passing == 'n':
    print(subject)
    print('다음 단계로 넘어갑니다.')
    break


#학생 점수 입력
student_number = 1
print('1명씩 학생들의 점수를 과목 순서대로 입력하세요.')
while int(student_number) < int(len(student))+1:
  for i in range(len(subject)):
    ss=input(student[int(student_number)-1]+'의 점수를 입력하세요.') 
    score[int(float(student_number)-1)].append(ss)
  print('다음 학생으로 넘어갑니다.')
  student_number=student_number+1
print(score)

#평가단계 입력
stage=[]
standard = input('학교에서 실시하는 평가단계는 몇 단계입니까? 숫자로 적어주십시오.')
while int(standard) > int(len(stage)) :
  response=input('가장 높은 단계명부터 입력하세요.')
  stage.append(response)
print(stage)

#각 과목의 평가단계별 기준점수 설정하기
all_question = []
for i in range(int(len(subject))):
  subq = input(subject[i]+'의 총 문제 수는? ')
  all_question.append(subq)
print(all_question)

gap=[]
for i in range(len(subject)):
  if int(all_question[i])//int(standard) >= 1:
    gap.append(int(all_question[i])//int(standard))
  if int(all_question[i])//int(standard) < 1:
    gap.append(1)
print(gap)

#맞은 개수에 따라 평가단계 구분하기
if int(standard) == 3 :
  for i in range(len(student)): 
    for p in range(len(subject)): 
      if int(all_question[p]) - int(gap[p]) <= int(score[i][p]) <= int(all_question[p]):
        result[i].append(stage[0])
      elif int(all_question[p]) - int(gap[p])*2 <= int(score[i][p]) <= int(all_question[p]) - int(gap[p]) -1:
        result[i].append(stage[1])                              
      elif int(score[i][p]) <= int(all_question[p]) - int(gap[p])*2 -1:
        result[i].append(stage[2])
if int(standard) == 4 :
  for i in range(len(student)): 
    for p in range(len(subject)): 
      if int(all_question[p]) - int(gap[p]) <= int(score[i][p]) <= int(all_question[p]):
        result[i].append(stage[0])
      elif int(all_question[p]) - int(gap[p])*2 <= int(score[i][p]) <= int(all_question[p]) - int(gap[p]) -1:
        result[i].append(stage[1])                              
      elif int(all_question[p]) - int(gap[p])*3 <= int(score[i][p]) <= int(all_question[p]) - int(gap[p])*2 -1:
        result[i].append(stage[2])
      elif int(score[i][p]) <= int(all_question[p]) - int(gap[p])*3 -1:
        result[i].append(stage[3])
if int(standard) == 5 :
  for i in range(len(student)): 
    for p in range(len(subject)): 
      if int(all_question[p]) - int(gap[p]) < int(score[i][p]) <= int(all_question[p]):
        result[i].append(stage[0])
      elif int(all_question[p]) - int(gap[p])*2 <= int(score[i][p]) <= int(all_question[p]) - int(gap[p]) -1:
        result[i].append(stage[1])                              
      elif int(all_question[p]) - int(gap[p])*3 <= int(score[i][p]) <= int(all_question[p]) - int(gap[p])*2 -1:
        result[i].append(stage[2])
      elif int(all_question[p]) - int(gap[p])*4 <= int(score[i][p]) <= int(all_question[p]) - int(gap[p])*3 -1:
        result[i].append(stage[3])
      elif int(score[i][p]) <= int(all_question[p]) - int(gap[p])*4 -1:
        result[i].append(stage[4])
print(result)

#표로 나타내기
pd.DataFrame(result, columns=subject, index=student)
```
